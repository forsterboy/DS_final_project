```python

country_array = energy_df.country.unique().tolist()
pattern = r'\b\w*countries\w*\b'
income_country_filter_list = []
for country in country_array:
    if len(re.findall(pattern, country)) != 0:
        income_country_filter_list.append(country) 
```

```python
pattern = r'\(\w*\)'
country_filter_list = []
for country in country_array:
    if len(re.findall(pattern, country)) != 0:
        country_filter_list.append(country)
```

```python
continent_filter_list = ['Europe']
continents_df = energy_df.query('country in @continent_filter_list')
```

```python
continent_filter_list = ['ASEAN ', 'Africa', 'Asia', 'Asia & Oceania', 'Asia Pacific', 'Europe', 'North America', 'South America']
```

```python
# Energy informations distributed to diff. income countries
energy_income_countries_df = energy_df.query('country in @income_country_filter_list')
# Grouped information from diff. organisations vs                                    
energy_grouped_countries_df = energy_df.query('country in @country_filter_list')
# Continental energy consumption information
energy_continental_df = energy_df.query('country in @continent_filter_list')

#
energy_countries_df = energy_df.query('country not in @income_country_filter_list and country not in @country_filter_list and country not in @continent_filter_list')
```

```python
print(energy_income_countries_df.shape)
print(energy_grouped_countries_df.shape)
print(energy_continental_df.shape)
print(energy_countries_df.shape)
print('Rows ', energy_income_countries_df.shape[0] + energy_grouped_countries_df.shape[0] + energy_continental_df.shape[0] + energy_countries_df.shape[0])
```
